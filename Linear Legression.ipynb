```python
data = {  
    "distance_km": ["50km", "100km", "150km", "200km", "250km", "300km", "350km", "400km", "450km", "500km"],  
    "fuel_liters": ["2.5lt", "5lt", "7.5lt", "10lt", "12.5lt", "", "17.5lt", "", "22.5lt", "25lt"]  
} 
```
